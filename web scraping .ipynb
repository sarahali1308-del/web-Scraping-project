```python
import pandas as pd
import requests
import io

# 1. Get the page with our "disguise"
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'
headers = {'User-Agent': 'Mozilla/5.0'}
response = requests.get(url, headers=headers)

# 2. Use io.StringIO so pandas doesn't look for a file on your disk
# This fixes the 'FileNotFoundError' from your photo!
mystr_input = io.StringIO(response.text)
df_list = pd.read_html(mystr_input)

# 3. Grab the table and save it
df = df_list[0]
df.to_csv('largest_companies.csv', index=False)

print("Success! Your file 'largest_companies.csv' is ready!")
df.head()
```
